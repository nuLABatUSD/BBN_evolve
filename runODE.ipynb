```python
import numpy as np
import matplotlib.pyplot as plt
import ODESolve_slow as ODE
```

Initial conditions: 
- x0 = 0 (initial independent variable)
- y0 = (1, 0, $- \pi^2$, 0) (initial dependent variables)
- dx0 = 0.1 (initial step-size guess)

The parameter p = ``None``. We don't need this here so we're just putting nothing. We will probably use ``p`` at another time, but for now we leave it becaues our functions require it.

Parameters for ``ODE.ODEOneRun``:
- ``N_step`` is the maximum number of steps for the ODE Solver
- ``dN`` is the number of ODE solving steps per step
- ``x_final`` is the final x value. The ODE solver will stop when it reaches this final x.

```python
x0 = 0
y0 = np.array([1, 0, - np.pi**2, 0])

dx0 = 0.1
p = None

N_step = 1000
dN = 1
x_final = 10
```

``ODE.ODEOneRun`` creates 4 outputs:
- x, an array of independent variables
- y, an array of arrays of independent variables (``y[:,0]``, ``y[:,1]``, etc)
- dx, an array of step sizes
- done, a true or false value to tell us if we reached ``x_final``

```python
x, y, dx, done = ODE.ODEOneRun(x0, y0, dx0, p, N_step, dN, x_final)
print(done)
```

We plot the solution. We can replace the ``0`` in ``y[:,0]`` with 1, 2, or 3, which will represent the other solutions.

```python
plt.figure()
plt.plot(x, y[:,0])
```
